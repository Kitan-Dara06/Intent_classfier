```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import LabelEncoder
```

```python
y_train = train_df_combined['label_name']
```

```python
X_train = train_df_combined['text']
```

```python
import re

def clean_text(text):
    text = text.lower()  # lowercase
    text = re.sub(r'[^\w\s]', '', text)  # remove punctuation
    return text.strip()  # remove leading/trailing whitespace

X_train_clean = [clean_text(text) for text in X_train]
```

```python
vectorizer = TfidfVectorizer()
X_train_vec = vectorizer.fit_transform(X_train_clean)
clf = LogisticRegression()
clf.fit(X_train_vec, y_train)

# Look at top features
import numpy as np

feature_names = np.array(vectorizer.get_feature_names_out())
for i, class_label in enumerate(clf.classes_[:5]):  # just show first 5 classes
    top10 = np.argsort(clf.coef_[i])[-10:]
    print(f"\nTop features for class '{class_label}':")
    print(feature_names[top10])
```

```python
%%writefile intent_app.py
import streamlit as st
import pickle

with open('intent_classifier.pkl', 'rb') as f:
    model = pickle.load(f)

st.title("🧠 Intent Classifier")
user_input = st.text_input("Enter your sentence:")
if user_input:
    prediction = model.predict([user_input])[0]
    st.success(f"Predicted Intent: {prediction}")
```
